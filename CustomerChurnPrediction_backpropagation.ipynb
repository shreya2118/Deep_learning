BINARY CLASSIFICATION PROBLEM--

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Churn_Modelling.csv")
df.head()
#df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df['Exited'].value_counts()
```

```python
df['Geography'].value_counts()
```

Removing first three columns

we have used "inplace" to make permanent changes

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df.head()
```

applying one-hot encoding to categorial columns

```python
df=pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
```

```python
df
```

For scaling, first perform train test split, and then standardize

```python
X=df.drop(columns=['Exited'])
y=df['Exited']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

Sequential model

By this we added our input and output layer

```python
model=Sequential()
#1 i/p layer, 1 hidden layer= 3 perceptrons/nodes, 1 o/p layer

model.add(Dense(3,activation='sigmoid',input_dim=11))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam')

#optimizer is used for gradient descent
```

```python
model.fit(X_train_scaled,y_train,epochs=10)
#model gets trained
```

```python
#weights ki value and bias=3 ki value
#input layer or next hidden layer ke weights h

model.layers[0].get_weights()
```

```python
#hidden layer or output layer ke beech ke weights and bias

model.layers[1].get_weights()
```

Now, the model is fully trained, and we can use it now for the prediction

```python
z=model.predict(X_test_scaled)
```

The values are not in 0/1 because we have used sigmoid function, we can change it to 0/1 by setting a **threshold** and then saying if it is greater than the threshold, output that as 1 or 0

```python
y_pred=np.where(z>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

**Right now, the accuracy is 0.8235, we can increase the accuracy by **

*  increasing Epochs
*  activation function set to "ReLu"
*  increase the number of nodes of hidden layers
*  increase the number of hidden layers




**CHANGED CODE**

```python
model=Sequential()
#1 i/p layer, 2 hidden layer= 11 perceptrons/nodes,, 1 o/p layer

model.add(Dense(11,activation='relu',input_dim=11))
model.add(Dense(11,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

KEEP IN MIND, THAT TRAING ACCURACY AS WELL AS VALIDATION ACCURACY BOTH SHOULD INCREASE,

---

IF VALIDATION ACCURACY IS NOT INCREASING, IT BECOMES THE CASE OF OVERFITTING

```python
model.fit(X_train_scaled,y_train,epochs=25,validation_split=0.2)
```

```python
model.layers[2].get_weights()
```

```python
z=model.predict(X_test_scaled)
```

```python
y_pred=np.where(z>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

**ACCURACY HAS BEEN IMPROVED TO** 0.862
