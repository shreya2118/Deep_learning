```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("perceptron_example.csv")
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(x='cgpa', y='resume_score', hue='placed', data=df)
```

using perceptron

extract x and y

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron(max_iter=1000,eta0=0.1,random_state=42)
```

training the perceptron

```python
p.fit(x,y)
```

```python
p.coef_
```

```python
p.intercept_
```

Standardize your data: as perceptrons are really sensitive to the features

```python
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_std=scaler.fit_transform(x)
p.fit(x_std,y)
```

```python
plot_decision_regions(x_std, y.values, clf=p)
```
