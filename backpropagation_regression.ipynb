```python
import numpy as np
import pandas as pd
```

Step 1: initialize parameters==, with this function: we will give them the dimension of number of neurons in input layer, hidden layer and output layer and then it will automatically generates the number of learning parametrs found.


```python
def initialize_parameters(layer_dims):
  np.random.seed(3)
  parameters={}
  L=len(layer_dims)
  for i in range(1,L):
    parameters['W'+str(i)]=np.ones((layer_dims[i-1],layer_dims[i]))*0.1
    parameters['b'+str(i)]=np.zeros((layer_dims[i],1))
  return parameters
```

Now, the "linear_forward" function calculates the output of every neuron present in the structure

```python
def linear_forward(A_prev,W,b):
  Z=np.dot(W.T,A_prev)+ b

  return Z
```

the main function which will perform the forward propagation

```python
def L_layer_forward(X,parameters):
  A=X
  L=len(parameters)//2                #number of layers in neural network
  for i in range(1,L+1):
    A_prev=A
    W1=parameters['W'+str(i)]
    b1=parameters['b'+str(i)]

    #print("A"+str(i-1)+': ',A_prev)
    #print("W"+str(i)+': ',W1)
    #print("b"+str(i)+': ',b1)
    #print("--"*20)

    A=linear_forward(A_prev,W1,b1)
    #print("A"+str(i)+': ',A)
    #print("**"*20)
  return A,A_prev
```

this is the step where we are predicting the output of every student

```python

X=df[['cgpa','score']].values[0].reshape(2,1)
y=df[['lpa']].values[0][0]

#if we print "X" it will give the two inputs of first student
#and if we print"y" it will give us the value of lpa(input)

#paramter initialization:
parameters=initialize_parameters([2,2,1])
L_layer_forward(X, parameters)
```

update paramters using gradient descent: these are all 9 formulaes of the derivatives we calculated

```python
def update_parameters(parameters,y,y_hat,A1,X):
  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (y - y_hat)*A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat)*A1[1][0])
  parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0])

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0])
```

```python
X=df[['cgpa','score']].values[0].reshape(2,1)
y=df[['lpa']].values[0][0]

#paramter initialization:
parameters=initialize_parameters([2,2,1])
y_hat,A1=L_layer_forward(X, parameters)
y_hat=y_hat[0][0]                                          # for 1st student

update_parameters(parameters,y,y_hat,A1,X)
parameters
```

```python
X=df[['cgpa','score']].values[1].reshape(2,1)
y=df[['lpa']].values[1][0]

y_hat,A1=L_layer_forward(X, parameters)
y_hat=y_hat[0][0]                                          # for 2nd student

update_parameters(parameters,y,y_hat,A1,X)
parameters
```

```python
X=df[['cgpa','score']].values[3].reshape(2,1)
y=df[['lpa']].values[3][0]

#paramter initialization:
parameters=initialize_parameters([2,2,1])
y_hat,A1=L_layer_forward(X, parameters)
y_hat=y_hat[0][0]                                          # for 4th student

update_parameters(parameters,y,y_hat,A1,X)
parameters
```

now, we write the for loop for the number of epochs to run

```python
#epochs implementation
parameters=initialize_parameters([2,2,1])

epochs=5
for i in range(epochs):
  loss=[]
  for j in range(df.shape[0]):
    X=df[['cgpa','score']].values[j].reshape(2,1)
    y=df[['lpa']].values[j][0]

    y_hat,A1=L_layer_forward(X, parameters)
    y_hat=y_hat[0][0]                                          # for j th student

    update_parameters(parameters,y,y_hat,A1,X)
    loss.append((y-y_hat)**2)
  print("Epoch- ",i+1,"Loss- ",np.array(loss).mean())

parameters
```

**CODE USING KERAS, TENSORFLOW**

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(2,activation='linear',input_dim=2))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

random initialization of weights

```python
model.get_weights()
```

```python
new_weights=[np.array([[0.1,0.1],[0.1,0.1]],dtype=np.float32),
             np.array([0.,0.],dtype=np.float32),
             np.array([[0.1],
                       [0.1]],dtype=np.float32),
             np.array([0.],dtype=np.float32)]
```

```python
model.set_weights(new_weights)
```

```python
model.get_weights()
```

```python
optimizer=keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='mean_squared_error',optimizer=optimizer)
```

```python
model.fit(df.iloc[:,0:-1].values,df['lpa'].values,epochs=75,batch_size=1,verbose=1)
```
