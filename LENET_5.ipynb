LENET-5 architecture consists of 5 layers.
2 convolutional + pooling layers,
2 Fully connected layers,
1 output layer

As the number of convolutional layers increases the number of filters also increases.

```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense, Flatten, Conv2D, AveragePooling2D
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
model=Sequential()
```

```python
model.add(Conv2D(6,kernel_size=(5,5),padding='valid',activation='tanh',input_shape=(32,32,1)))
model.add(AveragePooling2D(pool_size=(2,2),strides=2,padding='valid'))
model.add(Conv2D(16,kernel_size=(5,5),padding='valid',activation='tanh'))
model.add(AveragePooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Flatten())

model.add(Dense(120,activation='tanh'))
model.add(Dense(84,activation='tanh'))
model.add(Dense(10,activation='softmax'))
```

Pooling layer- does not contain any parameters,
Convolution layer- will include both weights and biases

For 1st convolution layer= no of learnable parameters: (kernel_size +1)* no of channels currently

For 2nd convolution layer= no of learnable parameters: (kernel_size*no.of channels previously +1) * no of channels currently
